Departed Pax for Nordics

Goal: Explore data and predict Online Bookings and departed Pax for 2019 for Thomas Cook Nordics
This notebook mainly focuses on the Time Series Analysis.

We start with the Exploratory Data Analysis of the main metrics revealing present trends and patterns in the data, giving a solid foundation for the further causal analysis.

I am utilising yearly data from Jan 2014 to Dec 2017 ,it would have been ideal to get hold of more historical data to understand trends but thats what we got in CMI table at present



Time series data is divided into the following three components:

trends: these are non-periodic and systematic trends in the data,
seasonal effects: these are modelled as daily or yearly periodicities in the data, and
one-off effects: these are effectively outliers.
Each of these components contribute additively/multiplicatively to the observed time series.The model can be mathematically written as follows:

y(t)=g(t)+s(t)+h(t)+ϵt

where y(t) is the Pax on day t and g(t),s(t) and h(t) correspond to the growth trend, seasonal effects and outliers respectively. Finally, ϵt accounts for noise.


We now discuss each of the components of the model

Trend component
The default trend component is a linear growth trend with some changepoints. If there are no changepoints, this means there willl be growth with some linear rate such that:

g(t)=αt

Changepoints are introduced so that the linear rate of growth can vary. For example, it may be the case that there was 2% daily growth (corresponding to g(t)=1.02×t) for the first year and 1% daily growth (so that g(t)=1.01×t) afterwards.

Seasonal components
Seasonality is a crucial component of time-series and we observe yearly and weekly seasonality at Thomas Cook

outliers
These are data points which deviate significantly from the distribution of the remainder of the data. If these points are ignored they may possibly lead to poor forecasts and so it is important to consider these points carefully. Examples may include extreme weather data or world cup football event that impacted travel.

```python
#importing holiday data for nordics region - only bank holidays taken into account
import google.datalab.bigquery as bq
httplogs_query = bq.Query('SELECT * FROM `lively-clover-826.rahul.holidays_nordics`')
output_options = bq.QueryOutput.table(use_cache=False)
result = httplogs_query.execute(output_options=output_options).result()
holidays=result.to_dataframe()
```

```python
holidays.trip_date_start = pd.to_datetime(holidays.trip_date_start)
holidays = holidays.rename(columns={'trip_date_start': 'ds'}) 
holidays.head()
```

```python
holidays['lower_window'] = holidays['lower_window'].map({-7: -21})
```

```python
holidays['upper_window'] = holidays['upper_window'].map({7: 21})
holidays.head()
```

```python
#running the forecasting algorithm 
from fbprophet import Prophet
forecast_model = Prophet(growth='linear', seasonality_mode='additive',weekly_seasonality=11,yearly_seasonality=11,
                         holidays_prior_scale=30.0,holidays=holidays)
forecast_model.fit(dat)
```

```python
#Predicting the algorithm future values
df_dates = forecast_model.make_future_dataframe(periods=730, 
                                                include_history=True)
model_predictions = forecast_model.predict( df_dates )

#plot_pred = forecast_model.plot( model_predictions )
#plt.figure(figsize = (36, 6))
#plt.legend(loc='best', fontsize=20)
#plot_pred
```

```python
#Looking at the individual components of the model 
forecast_model.plot_components( forecast_model.predict(df_dates), 
                               uncertainty=False )
```

Weekly seasonality exists on saturday,yearly seasonality exists around months of july and august,overall increasing trend observed

```python
#pulling in adjusted forecasted values from bigquery table to determine bookings revenue for 2018 and 2019
import pandas as pd
import google.datalab.bigquery as bq
httplogs_query = bq.Query('SELECT * FROM `lively-clover-826.rahul.forecasted_adjusted_pax_nordics_terese_financeadjusted`')
output_options = bq.QueryOutput.table(use_cache=False)
result = httplogs_query.execute(output_options=output_options).result()
q=result.to_dataframe()

melted = pd.melt(q, id_vars='ds')
melted['col'] = melted.variable.str.slice(1,4)
melted['Title'] = melted['col'].astype(str).astype(int)
melted.ds = pd.to_datetime(melted.ds)
melted['new_date']=melted.ds-pd.to_timedelta(melted["Title"], unit='D')
```

```python
p=melted.groupby('new_date')['value'].sum().reset_index()
p['year'] = p['new_date'].dt.year
p['month'] = p['new_date'].dt.month

p['avgr1'] = np.where(np.array(p['month'] == 9),(p['value']*1.25),p['value'])
p['avgr1'] = np.where(np.array(p['month'] == 10),(p['value']*1.35),p['avgr1'])
p['avgr1'] = np.where(np.array(p['new_date'] >= '2017-08-19 00:00:00') & (p['new_date'] <= '2017-08-31 00:00:00'),(p['value']*1.25),p['avgr1'])
p['avgr1'] = np.where(np.array(p['new_date'] >= '2017-04-22 00:00:00') & (p['new_date'] <= '2017-04-23 00:00:00'),(p['value']*2),p['avgr1'])
p['avgr1'] = np.where(np.array(p['new_date'] >= '2018-04-07 00:00:00') & (p['new_date'] <= '2018-04-08 00:00:00'),(p['value']*2),p['avgr1'])
p['avgr1'] = np.where(np.array(p['new_date'] >= '2019-04-27 00:00:00') & (p['new_date'] <= '2019-04-28 00:00:00'),(p['value']*2),p['avgr1'])
```
