# TTC Delay Analysis

I've personally experienced many painful delays while traveling to work on the TTC and finally motivated myself to understand the how and why they happen. The data was surprisingly easy to obtain as the Ontario government is pretty transparent on public transportation information. The data is available in the TTC Project folder and can also be found with the following link (https://www.toronto.ca/city-government/data-research-maps/open-data/open-data-catalogue/). Keep in mind that I've renamed the files as they were named inconsistently and it would make it easier to combine.

Now that we get a picture of what our tables look like, we will now combine all the data using regular expression to select the appropriate files. Note that this was why we renamed the files earlier. The result will be a table with a shape of (99479, 10).



```python
regex = re.compile(r'SubwaySRTLogs.*')
files = os.listdir('./data')
file = filter(regex.search, files)

all_data = pd.DataFrame()
for datafile in file:
    df = pd.read_excel('data/' +  datafile)
    all_data = all_data.append(df)
    
all_data.shape
#stringColumns = ['Day', 'Station', 'Code', 'Bound', 'Line', 'Vehicle']
#all_data[stringColumns] = all_data[stringColumns].astype(str) 
```

## Data Cleansing

Now that the data has been combined we will clean as much of it as we can before we split it into training and testing sets. We first look at how much data is missing in the resulting table and will decide on whether or not we throw it out or impute it. 

```python
plt.figure(figsize=(12,8))
sns.heatmap(all_data.isnull())
```

We end up seeing a lot of missing data for Bound and a few for Line so we will remove all of it and any data related to the SRT line. The reason being is because we want to focus on delays near the downtown core and the SRT data will only muddy our predictive model. There were also inconsistent naming conventions fouund in the Station, Line, and Bound columns so we will remove those as well.

# Feature Engineering

Now we start creating some features based on our knowledge around how the subway operates

```python
# This column will be a boolean value representing whether there is a delay or not. It will also be our y-value 
all_data['Delay'] = all_data['Min Delay'].apply(lambda x: x>0) * 1
```

# Exploratory Data Analysis

```python
sns.barplot(x =all_data.Delay.value_counts().index, y=all_data.Delay.value_counts())
plt.title('Ride Results')
plt.ylabel('Number of Occurences');
```

Data is balanced enough between delays and non-delays so no balancing techniques are required.

```python
drop_cols = ['Time', 'Day', 'Station', 'Code', 'Min Delay', 'Min Gap',
       'Bound', 'Line', 'Vehicle']
y = all_data.drop(drop_cols, axis='columns')
y = (y.groupby('Date')[['Delay']]
      .sum()
      #.set_index('Date')
      .resample('M')
      .mean()
      #.reset_index()
      )
y.head()
```

data was aggregated to show average delays in a day for a given month

```python
plt.plot(y)
plt.show()
```

There seems to be some trend that exists within the data so we will break it down to the base components and then proceed to detrend the data to achieve stationarity

```python
from pylab import rcParams
pd.plotting.register_matplotlib_converters()
rcParams['figure.figsize'] = (18, 8)
decomposition = sm.tsa.seasonal_decompose(y, model='additive')
fig = decomposition.plot()
plt.show()
```

# Testing for Stationarity

```python
# reverse transformation at the end  using:
#np.exp(time_2_diff + time_1_diff.shift(1).fillna(0)+ time_1.shift(1).fillna(0)).head()
time_1 = np.log(y.Delay)
time_1_diff = (time_1 - time_1.shift(1)).dropna()
time_2_diff = (time_1_diff - time_1_diff.shift(1)).dropna()
```

```python
result = adfuller(time_2_diff)
print('ADF Statistic: %f' % result[0])
print('p-value: %.10f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

The detrending method chosen was to take the log of the measured value and then take the difference twice with a lag of 1 day. Stationarity was confirmed with the dickey fuller test resulting in rejecting the null hypothesis that our data is not stationary. Transformed data can be seen below after achieving stationarity

```python
plt.plot(time_2_diff)
```

# AR Process - AutocorrelationFunction

```python
from statsmodels.tsa import stattools
acf_result = stattools.acf(time_2_diff, nlags=1000)
plt.plot(acf_result)
plt.axhline(y=0, linestyle='--')
plt.axhline(y=-1.96/np.sqrt(len(y)), linestyle='--')
plt.axhline(y=1.96/np.sqrt(len(y)), linestyle='--')
```

# MA Process - Partial autocorrelation function 

```python
pacf_result = stattools.pacf(time_2_diff, nlags=20)
plt.plot(pacf_result)
plt.axhline(y=0, linestyle='--')
plt.axhline(y=-1.96/np.sqrt(len(y)), linestyle='--')
plt.axhline(y=1.96/np.sqrt(len(y)), linestyle='--')
```

```python
def ma1(theta= .5, n=100):
    time_series = []
    error = np.random.randn(n)
    for period in range(1, n):
        time_series.append(error[period] + theta*error[period-1])
    return pd.Series(time_series[1:], index=range(1, n-1))

def ar1(phi = .9, n=1000, init=0):
    time_series = [ init]
    error = np.random.randn(n)
    for period in range(2, n):
        time_series.append(error[period] + phi*time_series[-1])
    return pd.Series(time_series[1:], index=range(1, n-1))
```

# Hyperparameter tuning for ARIMA

Akaike Information Critera (AIC) will be the chosen metric to measure the performance of the model. Note that the lower the AIC value, the better, not the lowest absolute value of AIC.

```python
%%capture
p = d = q = range(0, 3)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

best_score = 100000

for param in pdq:
    #sleep(0.01)
    for param_seasonal in seasonal_pdq:

        try:
            mod = sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit()
            if results.aic < best_score:
                    best_score = results.aic
                    best_params = param, param_seasonal
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
print('The best parameters and score is:', best_params, best_score)
#ARIMA(0, 0, 0)x(1, 1, 0, 12)12 - AIC:-1.5092857497901973       
```

```python
pred = results.get_prediction(start=pd.to_datetime('2018-01-31'), dynamic=False)
pred_ci = pred.conf_int()
ax = y.Delay['2014':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Delays')
plt.legend()
plt.show()
```

```python
y_forecasted = pred.predicted_mean
y_truth = y['2018-01-31':]
mse = ((np.subtract(pd.DataFrame(y_forecasted), y_truth)) ** 2).mean()
print('The Root Mean Squared Error of our forecasts is:', round(np.sqrt(mse), 2)[0])
```

# Modeling with FB Prophet

```python
df2 = y.reset_index()
df2.columns = ['ds', 'y']
df2['y_orig'] = df2['y'] # to save a copy of the original data..you'll see why shortly. 
# log-transform y
df2['y'] = np.log(df2['y'])
df2.head()
```

```python
model = Prophet() #instantiate Prophet
model.fit(df2); #fit the model with your dataframe
```

```python
future_data = model.make_future_dataframe(periods=6, freq = 'm')
forecast_data = model.predict(future_data)
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
model.plot_components(forecast_data);
```

```python
df2 = df2.set_index('ds')
forecast_data = forecast_data.set_index('ds')
viz_df = df2.join(forecast_data[['yhat', 'yhat_lower','yhat_upper']], how='outer')
viz_df['yhat_rescaled'] = np.exp(viz_df.yhat)
viz_df.head()
```

```python
viz_df[['y_orig', 'yhat_rescaled']].plot()
```

```python
fig, ax = plt.subplots()
ax.plot(viz_df.y_orig)
ax.plot(viz_df.loc[viz_df.index > df2.index[-2], 'yhat_rescaled'], color='black', linestyle=':')
ax.fill_between(viz_df.index, 
                 np.exp(viz_df['yhat_upper']),
                 np.exp(viz_df['yhat_lower']),
                 alpha=0.5,
                 color='darkgray'
                )
ax.set_title('Delay (Orange) vs Delay Forecast (Black)')
ax.set_ylabel('Average Daily Delay')
ax.set_xlabel('Date')
```

```python
y_forecasted2 = viz_df.loc['2018-01-31':'2018-10-31', 'y_orig']
y_truth2 = viz_df.loc['2018-01-31':'2018-10-31', 'yhat_rescaled']
mse = ((np.subtract(y_forecasted2, y_truth2)) ** 2).mean()
print('The Root Mean Squared Error of our forecasts is:', round(np.sqrt(mse), 2))
```
